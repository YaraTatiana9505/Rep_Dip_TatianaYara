##2. **Redes neuronales convoluciones**<a id="Redes neuronales convoluciones"></a>

Las redes neuronales convolucionales (CNN, Convolutional neural network) son una
familia de modelos inspirados en cómo trabaja la corteza visual del cerebro humano
en el reconocimiento de objetos. La estructura de una CNN "clasica" se compone de una o
más capas convulocionales, seguidas de una capa de agrupamiento y una capa completamente
conectada.

El primer desarrollo de las CNN empieza en los años noventa,
con el articulo "Handwritten Digit Recognition with a Back-Propagation Network"
 de LeCun y otros, en 1989
  (https://papers.nips.cc/paper/1989/file/53c3bce66e43be4f209556518c2fcb54-Paper.pdf).

  En estos últimos años se han popularizado enormemente al conseguir resultados impresionantes en el reconocimiento de imagenes. Existe un rasgo diferencial en las  redes neuronales convolucionales es que hacen la suposición explicita de que la entrada son imagenes, lo cual permite codificar ciertas propiedades en la arquitectura para reconocer elementos concretos en las imágenes.

##3. **Idea Intuitiva de una CNN**<a id="Idea Intuitiva de una CNN"></a>

La clave de cualquier algoritmo de aprendizaje automatico es la estracción de características relevante.
Las redes neuronales son capaces de aprender automáticamente las características a partir de datos
primarios que son más útiles para una tarea particular; por ello, las primeras capas extraen **características de bajo
nivel**.

Por ejemplo, si estamos trabajando con imagenes, las características de bajo nivel
(como bordes y manchas) se extraen de las primeras capas, las cuales se combinan para
formar características de alto nivel, como las formas de los objetos (edificios, coches, personas, etc). Es decir, en una CNN cada capa va aprendiendo diferentes niveles de abstracción.

##4. **Componentes básicos de una CNN**<a id="Componentes básicos de una CNNs"></a>

###4.1 **Operación de convolución**<a id="Operación de convolución"></a>
La diferencia entre una capa densa y una capa especializada en la operación de convolución (Capa convolucional), es que la capa densa aprende patrones globales en su espacio global de entrada, mientras que la capa convolucional aprende patrones locales dentro de la imagen en pequeñas ventanas de dos dimensiones. 

<center><img src=https://i1.wp.com/datasmarts.net/es/wp-content/uploads/2018/11/full_padding_no_strides_transposed.gif?resize=264%2C300&ssl=1 width="200"><center>  

Para este objetivo se requieren los siguientes hiperparametros:

####4.1.1 **Tamaño**<a id="Tamaño"></a>
En general las capas convoluciones operan sobre tensores 3D llamados feature maps, con dos ejes espaciales (alto y ancho) y un eje de canal, también llamado profundidad.

####4.1.2 **Stride**<a id="Stride"></a>
En este caso no se conectan todas las neuronas de entrada con todas las neuronas del primer nivel si no que se hace por pequeñas zonas localizadas  de los pixeles de la imagen. Se pueden hacer movimientos avanzando de a un determinado número de pixeles (parámetro llamado stride).

####4.1.3 **Padding**<a id="Padding"></a>
Además en CNN también se puede aplicar una técnica de relleno de ceros alrededor del margen de la imagen, el parametro para definir este relleno recibe el nombre de padding. Dado que a veces se quiere obtener un tensor de salida de las mismas dimensiones de la entrada.

<center><img src=http://torres.ai/wp-content/uploads/2018/06/Picture.4.9.png width="500"><center> 

### 4.2 **Operación de pooling**<a id="Operación de pooling"></a>
Capa de agrupación, suele ser aplicada inmediatamente después de las capas convolucionales. Estás capas hacen una simplificaciónde la información recogida por la capa convolucional. Una manera condensar la información es utilizando max-pooling, es decir se toma el máximo valor por una ventana de tamaño definido, por ejemplo 2x2.

<center><img src=https://i1.wp.com/www.aprendemachinelearning.com/wp-content/uploads/2018/11/cnn-05.png width="400"><center> 

### 4.3 **Capas de Dropout**<a id="Capas de Dropout"></a>
Técnica que ignora ciertos conjuntos de neuronas de la red neuronal durante la fase de entrenamiento de manera aleatoria de tal manera se mitiga el sobreajuste.
### 4.4 **BatchNormalization**<a id="BatchNormalization"></a>
Normaliza las entradas de la capa de tal manera que tenga una activación de salida media cero y desviación estándar uno.

##5. **Detección de objetos**<a id="Detección_de_objetos"></a>
Se describen dos enfoques más comunes en la detección de objetos 2D para imágenes RGB, R-CNN y YOLO.

###5.1 **MASK-RCNN for RGB Images**<a id="R-CNN_for_RGB_Images"></a>
El modelo de red neuronal convolucional basada en la región de máscara, o máscara R-CNN, es uno de los enfoques más avanzados para las tareas de reconocimiento de objetos. Existen cinco variaciones principales del enfoque, lo que da como resultado el método actual llamado Mask R-CNN. Los aspectos más destacados de cada variación se pueden resumir de la siguiente manera:

#### 5.1.1 **CNN**
Las CNN han mostrado resultados impresionantes en el reconocimiento de imágenes y se han convertido en el modelo estándar para las tareas de clasificación de imágenes.
La detección de objetos en una imagen, utilizando CNN, se puede resumir en tres pasos: 

(a) Crear un conjunto de cuadros delimitadores (ancho, alto y centro del cuadro).

(b) Mover cada cuadro delimitador en distintas direcciones x e y.

(c) En cada movimiento, la imagen en el cuadro delimitador se pasa como entrada a un clasificador, una CNN para predecir si el cuadro delimitador contiene un objeto y, de ser así, qué objeto. Este enfoque es ineficaz y no se puede utilizar en tiempo real debido al costoso tiempo de cálculo.

#### 5.1.2 **R-CNN**
La red neuronal convolucional basada en regiones, o R-CNN, es una familia de modelos de redes neuronales convolucionales diseñados para la detección de objetos, desarrollado por Ross Girshick.
En lugar de clasificar todas las regiones posibles en una imagen, utiliza un algoritmo de búsqueda selectiva que genera un pequeño conjunto de regiones representadas como cuadros delimitadores que pueden contener un objeto. 

R-CNN funciona bien, pero sigue siendo ineficaz, ya que tiene que aprobar cada región propuesta para cada imagen en un CNN y clasificarlos usando un clasificador SVM. Se utiliza una CNN previamente entrenada para extraer características de la imagen y para determinar si la región propuesta contiene un objeto o no.

#### 5.1.3 **Fast R-CNN** 

Este modelo en lugar de utilizar una CNN previamente entrenada para extraer funciones para cada región, permite ejecutar CNN solo una vez para todas las propuestas de región en una imagen de entrada.
Esto se hace mediante una operación conocida como agrupación de regiones de interés (RoIPooling), que proyecta cada región propuesta (cuadro delimitador) desde la imagen de entrada a un área en la última capa de convolución (también conocida como mapas de características) y luego aplica la agrupación máxima para cada región. La arquitectura de Fast R-CNN viene con una única CNN entrenable para la extracción y clasificación de características, donde RCNN tiene modelos separados para la extracción de características y la clasificación de objetos.

#### 5.1.4 **Faster R-CNN**
La arquitectura de Faster R-CNN es una única tubería entrenable que no requiere un algoritmo para las propuestas de región. El algoritmo de búsqueda de selección utilizado en R-CNN y Fast R-CNN se reemplaza por una pequeña red convencional denominada Region Proposal Network (RPN) que genera regiones de propuesta.
Específicamente, el algoritmo tiene en cuenta los siguientes pasos:

* Primero usa una ConvNet para extraer mapas de características de las imágenes.

* Estos mapas de características luego se pasan a través de una red de propuesta de región (RPN) que devuelve los cuadros delimitadores candidatos.

* Luego aplicamos una capa de agrupación de RoI en estos cuadros delimitadores candidatos para que todos los candidatos tengan el mismo tamaño

* Y finalmente, las propuestas se pasan a una capa completamente conectada para clasificar y generar los cuadros delimitadores de los objetos.

#### **Mask R-CNN**
Mask R-CNN es básicamente una extensión de Faster R-CNN. Mask R-CNN es la variación más reciente de la familia de modelos y admite tanto la detección de objetos como la segmentación de objetos. La segmentación de objetos no solo implica localizar objetos en la imagen, sino que también especifica una máscara para la imagen, que indica exactamente qué píxeles de la imagen pertenecen al objeto. Para una imagen determinada, devuelve la etiqueta de clase y las coordenadas de la máscara para cada objeto de la imagen. Tiene en cuenta los siguientes componentes:

* Modelo de columna vertebral

De manera similar a la ConvNet que usamos en Faster R-CNN para extraer mapas de características de la imagen. Entonces, el primer paso es tomar una imagen y extraer características usando la arquitectura ResNet 101. Estas características actúan como una entrada para la siguiente capa.

* Red de propuestas regionales (RPN)

Ahora, tomamos los mapas de características obtenidos en el paso anterior y aplicamos una red de propuesta de región (RPM). Básicamente, esto predice si un objeto está presente en esa región (o no). En este paso, obtenemos aquellas regiones o mapas de características que el modelo predice que contienen algún objeto.

* Región de interés (RoI)

Las regiones obtenidas del RPN pueden tener diferentes formas. Por lo tanto, aplicamos una capa de agrupación y convertimos todas las regiones a la misma forma. A continuación, estas regiones se pasan a través de una red completamente conectada para predecir la etiqueta de clase y los cuadros delimitadores.

Hasta este punto, los pasos son casi similares a cómo funciona Faster R-CNN. Ahora viene la diferencia entre los dos marcos. Además de esto, Mask R-CNN también genera la máscara de segmentación.

Para eso, primero se calcula la región de interés para poder reducir el tiempo de cálculo. Para todas las regiones predichas, calculamos la intersección sobre la unión (IoU) con los cuadros de verdad del terreno. Podemos computar IoU de esta manera:

IoU = Área de la intersección / Área de la unión

Ahora, solo si el IoU es mayor o igual a 0.5, lo consideramos como una región de interés. De lo contrario, se descarta esa región en particular. Hacemos esto para todas las regiones y luego seleccionamos solo un conjunto de regiones para las cuales el IoU es mayor que 0.5.

###5.2 **YOLO para Imagenes RGB**<a id="YOLO_para_Imagenes_RGB"></a>

You Only Look Once (YOLO) es un enfoque alternativo para la detección de objetos. La arquitectura de YOLO es una CNN simple, que se alimenta de una imagen solo una vez a través de la red y genera las puntuaciones de la clase y las coordenadas del cuadro delimitador. 

####5.2.1 **YOLO V1**<a id="YOLOv1"></a>
YOLO divide la imagen de entrada en una cuadrícula S × S, y cada celda de la cuadrícula predice B cuadros delimitadores con puntajes de confianza y puntajes de clase. 


<center><img src=https://sheng-fang.github.io/img/post_img/2020-04-25-review_yolo/yolo_grid.png width="300"><center> 

Las regiones de objetos (el rectángulo verde) en la imagen se representan como un cuadro. El centro de la caja (el punto verde) se define como el centro de un objeto. A medida que la imagen se divide en varias celdas de la cuadrícula, el centro de un objeto cae en una celda de la cuadrícula. Consideramos esta celda de la cuadrícula como el predictor, que predice la clase de objeto y las cuatro esquinas del cuadro.

#####5.2.1.1 **Arquitectura**<a id="Arquitectura"></a>
<center><img src=https://sheng-fang.github.io/img/post_img/2020-04-25-review_yolo/yolo_arch.png width="600"><center> 

Durante la implementación, la última capa completamente conectada toma un vector de 4096 como entrada y salidas un vector de 7 x 7 x 30. 

En YOLO V1:

Sin capa de normalización por lotes.

Función de activación: LeakyReLU para todas las capas, excepto la última, que utiliza activación lineal.

Dropout: capa con tasa = 0,5 después de la primera capa completamente conectada

####5.2.1.3 **Beneficios**<a id="Beneficios"></a>

-Rápido
-El sistema de entrenamiento global facilita la optimización
-Más generalizado, cuando se prueba en otra base de datos

####5.2.1.4 **Inconvenientes**<a id="Inconvenientes"></a>

-Mal rendimiento cuando hay grupos de objetos pequeños, porque cada cuadrícula solo puede detectar un objeto.
-El error principal de YOLO proviene de la localización, porque la relación del cuadro delimitador se aprende totalmente a partir de los datos y YOLO comete un error en el cuadro delimitador de relación inusual.

####5.2.2 **YOLO V2**<a id="YOLOv1"></a>
 Para remediar los 2 problemas mencionados anteriormente y mejorar el rendimiento, los autores realizan varias modificaciones en YOLO V2.


#####5.2.2.1 **Normalización por lotes**<a id="Normalización_por_lotes"></a>

Agregar la normalización por lotes ayuda a mejorar mAP en un 2%.

#####5.2.2.2 **Clasificador de alta resolución**<a id="Clasificador_de_alta_resolución"></a>
YOLO V2 está entrenado en imágenes de alta resolución durante las primeras 10 épocas. Esto mejora la mAP en un 4%.

#####5.2.2.3 **Convolucional con cajas de anclaje**<a id="Convolucional_con_cajas_de_anclaje"></a>
Uno de los inconvenientes de YOLO V1 es el mal rendimiento en la localización de cuadros, porque los cuadros delimitadores aprenden totalmente de los datos. En YOLO V2, los autores agregan anteriores (cuadros de anclaje) para ayudar a la localización. Para introducir los anclajes, se realizan algunas modificaciones en la arquitectura de la red.

1. Elimina las capas completamente conectadas
2. Reduzca la dimensión de la imagen de entrada.
3. Mueva la predicción de la clase de objeto de la celda de la cuadrícula al cuadro delimitador. Esto significa que una celda puede predecir diferentes objetos con diferentes cajas. 

#####5.2.2.4 **Conglomerados de dimensiones**<a id="Conglomerados_de_dimensiones"></a>
Este proceso decide el número de anclajes y las proporciones predefinidas para los anclajes.

#####5.2.2.5 **Predicción de ubicación directa**<a id="Predicción_de_ubicación_directa"></a>
Los valores de las coordenadas del centro de la caja, X y y, están entre 0 y 1. Pero YOLO V1 puede predecir los valores menores que 0 o mayores que 1. Los autores agregan activación logística para forzar que la predicción caiga dentro de este rango.

#####5.2.2.6 **Característica Fine-Grained**<a id="Característica_Fine-Grained"></a>
Esta es una idea similar a la pirámide de características. Los autores remodelan los últimos mapas de características con una resolución de 26 x 26 a 13 x 13, y concatenan los mapas de características remodelados a la salida de la última capa convolucional. La última capa convolucional genera un tensor de 13x13x1024. El mapa de características remodelado tiene una dimensión de 13x13x2048. Los resultados de concatenación tienen una dimensión de 13x13x3072.

#####5.2.2.7 **Entrenamiento multiescala**<a id="Entrenamiento_multiescala"></a>
Entrenando el modelo con imágenes de diferentes dimensiones. La dimensión de las imágenes se elige aleatoriamente cada 10 lotes durante el entrenamiento.

####5.2.3 **YOLO V3**<a id="YOLOv3"></a>
Los autores de YOLO han probado muchas técnicas para mejorar la precisión y el rendimiento. Cuatro de ellos funcionan bien para YOLO.


#####5.2.3.1 **Predicción del cuadro delimitador**<a id="Predicción_del_cuadro_delimitador"></a>
YOLO V2 considera la puntuación de confianza como la multiplicación de $p (Object)$ y $IOU(b,Object )$. Primero calculamos la probabilidad de objetividad, luego tomamos en cuenta el pagaré para ajustar la probabilidad. Mientras que YOLO V3 proporciona una comprensión más intuitiva, que considera la puntuación de confianza de la caja directamente como la probabilidad de objetividad. Entonces, la probabilidad de objetividad se define como el IOU de la caja predicha y la verdad fundamental. Al usar la activación logística, la predicción se fuerza al rango entre 0 y 1.

#####5.2.3.2 **Predicción de clases**<a id="Predicción de clases"></a> 
YOLO V2 usa softmax al calcular las probabilidades condicionales de clases. La técnica softmax impone la suposición de que cada caja tiene una sola etiqueta. Pero en realidad, a menudo le damos a los objetos varias etiquetas, que pueden superponerse. Para resolver el problema de múltiples etiquetas, YOLO V3 utiliza una clasificación logística independiente para cada clase. La entropía cruzada binaria se utiliza para calcular la pérdida de cada clase.

#####5.2.3.3 **Predicción a través de escalas**<a id="Predicción a través de escalas"></a> 
YOLO V3 predice 3 casillas en cada escala, por lo que el tensor es Sx Sx [ 3 x ( 4 + 1 + 80 ) ]. En general, las 3 escalas son:

- Última capa convolucional, que tiene 32 zancadas en comparación con la dimensión de entrada.
- La capa con 16 zancadas
- La capa con 8 zancadas
Si tenemos imágenes de entrada con una dimensión de 416 x 416, las 3 escalas tienen una dimensión de 13 x 13, 26 x 26 y 52 x 52.

```python
def draw_boxes(image_name):
    selected_value = full_labels[full_labels.filename == image_name]
    img = cv2.imread('gdrive/My Drive/Colab Notebooks/data/Kangaroo/images/{}'.format(image_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```
